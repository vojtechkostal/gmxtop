```python
from dataclasses import dataclass
from typing import Any, List, Tuple, Callable

@dataclass(frozen=True)
class Params:
    names: Tuple[str, ...]
    types: Tuple[type, ...]
    desc: str = ""


@dataclass(frozen=True)
class ParamBuilder:
    n_atoms: int
    funcs: dict[int, callable]

    def parse(self, func: int, param_tokens: List[str], ctx: str = "") -> dict:
        if func not in self.funcs:
            allowed = ", ".join(map(str, self.funcs.keys()))
            raise NotImplementedError(
                f"Only function(s) {allowed} are supported in {ctx}: function {func} is not recognized."
            )

        param_spec = self.funcs[func]

        params = {}
        for name, ptype, token in zip(param_spec.names, param_spec.types, param_tokens):
            try:
                params[name] = ptype(token)
            except ValueError as e:
                raise ValueError(
                    f"Invalid parameter value for '{name}' (expected type {ptype.__name__}) "
                    f"in function {func} of {ctx}: got '{token}'."
                ) from e

        return params


BONDS = ParamBuilder(
    n_atoms=2,
    funcs={
        1: Params(names=("b0", "kb"), types=(float, float), desc="harmonic bond"),
        2: Params(names=("b0", "kb"), types=(float, float), desc="G96 bond"),
        3: Params(names=("b0", "D", "beta"), types=(float, float, float), desc="Morse bond"),
        4: Params(names=("b0", "C2", "C3"), types=(float, float, float), desc="cubic bond"),
        5: Params(names=(), types=(), desc="connection (no parameters)"),
        6: Params(names=("b0", "kb"), types=(float, float), desc="harmonic potential"),
        7: Params(names=("bm", "kb"), types=(float, float), desc="FENE bond"),
        8: Params(names=("table", "k"), types=(int, float), desc="tabulated bond"),
        9: Params(names=("table", "k"), types=(int, float), desc="tabulated bond (variant)"),
        10: Params(names=("low", "up1", "up2", "kdr"), types=(float, float, float, float), desc="restraint"),
    },
)

def resolve_parts(
    parts: List[str],
    builder: ParamBuilder,
    section: str,
) -> Tuple[List, int, List[str]]:
    n_atoms = builder.n_atoms
    atom_tokens = parts[:n_atoms]
    param_tokens = parts[n_atoms + 1 :]  # skip function id
    func = int(parts[n_atoms])

    atoms = atom_tokens

    params = builder.parse(func, param_tokens, ctx=f"{section} (func={func})")

    return atoms, func, params
```

```python

from __future__ import annotations

from dataclasses import dataclass, field
from pathlib import Path
from typing import Any, Dict, List, Optional, Tuple, Union


# -----------------------------
# Data model
# -----------------------------

@dataclass(slots=True)
class Defaults:
    nbfunc: int
    comb_rule: int
    gen_pairs: str
    fudgeLJ: float
    fudgeQQ: float


@dataclass(slots=True)
class AtomType:
    name: str
    atnum: int
    mass: float
    charge: float
    ptype: str
    sigma: float
    epsilon: float


@dataclass(slots=True)
class BondType:
    ai: AtomType
    aj: AtomType
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class PairType:
    ai: AtomType
    aj: AtomType
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class AngleType:
    ai: AtomType
    aj: AtomType
    ak: AtomType
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class DihedralType:
    ai: AtomType
    aj: AtomType
    ak: AtomType
    al: AtomType
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class Atom:
    nr: int
    type: str
    resnr: int
    residue: str
    atom: str
    cgnr: int
    charge: float
    mass: Optional[float] = None


@dataclass(slots=True)
class Bond:
    ai: Atom
    aj: Atom
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class Pair:
    ai: Atom
    aj: Atom
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class Angle:
    ai: Atom
    aj: Atom
    ak: Atom
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class Dihedral:
    ai: Atom
    aj: Atom
    ak: Atom
    al: Atom
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class Exclusion:
    excluded: Tuple[int, ...]  # variable length per line


@dataclass(slots=True)
class Settle:
    ai: int
    func: int
    doh: float
    dhh: float


@dataclass(slots=True)
class VirtualSite3:
    a: int
    b: int
    c: int
    d: int
    func: int
    params: Dict[str, Any] = field(default_factory=dict)


@dataclass(slots=True)
class MoleculeType:
    name: str
    nrexcl: int
    atoms: List[Atom] = field(default_factory=list)
    bonds: List[Bond] = field(default_factory=list)
    pairs: List[Pair] = field(default_factory=list)
    angles: List[Angle] = field(default_factory=list)
    dihedrals: List[Dihedral] = field(default_factory=list)
    exclusions: List[Exclusion] = field(default_factory=list)
    settles: List[Settle] = field(default_factory=list)
    virtual_sites3: List[VirtualSite3] = field(default_factory=list)

    def __repr__(self):
        return f"<MoleculeType name={self.name}>"


@dataclass(slots=True)
class Topology:
    source: Path
    defaults: Optional[Defaults] = None
    atomtypes: List[AtomType] = field(default_factory=list)
    bondtypes: List[BondType] = field(default_factory=list)
    pairtypes: List[Pair] = field(default_factory=list)
    angletypes: List[Angle] = field(default_factory=list)
    dihedraltypes: List[Dihedral] = field(default_factory=list)
    moleculetypes: Dict[str, MoleculeType] = field(default_factory=dict)

    system: Optional[str] = None
    # molecules: List[Tuple[str, int]] = field(default_factory=list)  # (molname, count)
    molecules: Dict[MoleculeType, int] = field(default_factory=dict)

    def get_molecule(self, name: str) -> MoleculeType:
        return self.moleculetypes[name]


# -----------------------------
# Parsing utilities
# -----------------------------

def _strip_comment(line: str) -> str:
    # GROMACS uses ';' for comments; drop inline comments.
    # This is safe for typical .top/.itp (no quoted ';' usage in data lines).
    return line.split(";", 1)[0].strip()


def _try_number(tok: str) -> Any:
    # Convert to int/float where possible; otherwise return original string.
    try:
        # handle explicit ints (including negatives)
        if tok.lstrip("+-").isdigit():
            return int(tok)
    except Exception:
        pass
    try:
        return float(tok)
    except Exception:
        return tok


def _require(parts: List[str], n: int, *, section: str, source: Path, raw: str) -> None:
    if len(parts) < n:
        raise ValueError(f"Invalid [ {section} ] line in {source}: {raw}")


def _get_atom_by_nr(mol: MoleculeType, nr: int) -> Atom:
    return next((a for a in mol.atoms if a.nr == nr), None)


def _get_atomtype_by_name(top: Topology, name: str) -> AtomType:
        if name == "X":
            at = AtomType(name="X", atnum=0, mass=0.0, charge=0.0, ptype="A", sigma=0.0, epsilon=0.0)
            return at
        at = next((a for a in top.atomtypes if a.name == name), None)
        return at


def _match_wildcard(pattern: list[str, str, str, str], target: list[str, str, str, str]) -> bool:
    return all(p == "X" or p == t for p, t in zip(pattern, target))


def _lookup_paramtype(top: Topology, *atoms, pair=False) -> BondType | PairType | AngleType | DihedralType:

    at_forward = [a.type if isinstance(a, Atom) else a for a in atoms]
    at_reverse = list(reversed(at_forward))
    if len(atoms) == 2:
        if pair:
            for pt in top.pairtypes:
                pt_list = [pt.ai.name, pt.aj.name]
                if pt_list == at_forward or pt_list == at_reverse:
                    return pt
        else:
            for bt in top.bondtypes:
                bt_list = [bt.ai.name, bt.aj.name]
                if bt_list == at_forward or bt_list == at_reverse:
                    return bt


    elif len(atoms) == 3:
        for at in top.angletypes:
            at_list = [at.ai.name, at.aj.name, at.ak.name]
            if at_list == at_forward or at_list == at_reverse:
                return at
            
    elif len(atoms) == 4:
        best = None
        best_specifity = -1  # higher is better
        for dt in top.dihedraltypes:
            dt_list = (dt.ai.name, dt.aj.name, dt.ak.name, dt.al.name)

            # First, try exact match
            if dt == at_forward or dt == at_reverse:
                return dt
            
            # Next, try wildcard match (using 'X' as wildcard)
            if _match_wildcard(dt_list, at_forward) or _match_wildcard(dt_list, at_reverse):
                # Count number of specific (non-wildcard) entries
                specifity = sum(1 for p in dt_list if p != "X")
                if specifity > best_specifity:
                    best = dt
                    best_specifity = specifity

        return best
    return None


def _parse_atomtype(parts: List[str], top: Topology) -> AtomType:

    raw_line = " ".join(parts)
    section = "atomtypes"

    _require(parts, 7, section=section, source=top.source, raw=raw_line)
    name = parts[0]
    atnum = int(parts[1])
    mass = float(parts[2])
    charge = float(parts[3])
    ptype = parts[4]
    sigma = float(parts[5])
    epsilon = float(parts[6])
    return AtomType(name, atnum, mass, charge, ptype, sigma, epsilon)

    
def _parse_bondtype(parts: List[str], top: Topology) -> BondType:

    raw_line = " ".join(parts)
    section = "bondtypes"

    _require(parts, 3, section=section, source=top.source, raw=raw_line)
    ai = parts[0]
    aj = parts[1]
    funct = int(parts[2])
    if funct != 1:
        raise NotImplementedError(f"Only bond function 1 is supported in {section} of {top.source}: {raw_line}")

    _require(parts, 5, section="bondtypes", source=top.source, raw=raw_line)
    at_i = _get_atomtype_by_name(top, ai)
    at_j = _get_atomtype_by_name(top, aj)
    funct = int(parts[2])
    params = {'b0': float(parts[3]), 'kb': float(parts[4])}
    return BondType(at_i, at_j, funct, params)


def _parse_pairtype(parts: List[str], top: Topology) -> PairType:

    raw_line = " ".join(parts)
    section = "pairtypes"

    _require(parts, 3, section=section, source=top.source, raw=raw_line)
    ai = _get_atomtype_by_name(top, parts[0])
    aj = _get_atomtype_by_name(top, parts[1])
    funct = int(parts[2])

    if funct == 1:
        _require(parts, 5, section="pairtypes", source=top.source, raw=raw_line)
        params = {'sigma': float(parts[3]), 'epsilon': float(parts[4])}
    elif funct == 2:
        _require(parts, 8, section="pairtypes", source=top.source, raw=raw_line)
        params = {
            'fudgeQQ': float(parts[3]),
            'q1': float(parts[4]),
            'q2': float(parts[5]),
            'sigma': float(parts[6]),
            'epsilon': float(parts[7]),
        }
    else:
        raise NotImplementedError(f"Only pair function 1 and 2 are supported in {section} of {top.source}: {raw_line}")

    return PairType(ai, aj, funct, params)


def _parse_angletype(parts: List[str], top: Topology) -> AngleType:

    raw_line = " ".join(parts)
    section = "angletypes"

    _require(parts, 4, section=section, source=top.source, raw=raw_line)
    ai = _get_atomtype_by_name(top, parts[0])
    aj = _get_atomtype_by_name(top, parts[1])
    ak = _get_atomtype_by_name(top, parts[2])
    funct = int(parts[3])
    if funct == 1:
        _require(parts, 6, section="angletypes", source=top.source, raw=raw_line)
        params = {'th0': float(parts[4]), 'kth': float(parts[5])}
    elif funct == 5:
        _require(parts, 8, section="angletypes", source=top.source, raw=raw_line)
        params = {
            'th0': float(parts[4]),
            'kth': float(parts[5]),
            's0': float(parts[6]),
            'kub': float(parts[7]),
        }
    else:
        raise NotImplementedError(f"Only angle function 1 and 5 are supported in {section} of {top.source}: {raw_line}")

    return AngleType(ai, aj, ak, funct, params)


def _parse_dihedraltype(parts: List[str], top: Topology) -> DihedralType:

    raw_line = " ".join(parts)
    section = "dihedraltypes"

    _require(parts, 5, section=section, source=top.source, raw=raw_line)
    ai = _get_atomtype_by_name(top, parts[0])
    aj = _get_atomtype_by_name(top, parts[1])
    ak = _get_atomtype_by_name(top, parts[2])
    al = _get_atomtype_by_name(top, parts[3])
    funct = int(parts[4])

    if funct == 2:
        _require(parts, 7, section="dihedraltypes", source=top.source, raw=raw_line)
        params = {'xi': float(parts[5]), 'kxi': float(parts[6])}
    elif funct == 9:
        _require(parts, 8, section="dihedraltypes", source=top.source, raw=raw_line)
        params = {'phi0': float(parts[5]), 'kphi': float(parts[6]), 'mult': int(parts[7])}
    else:
        raise NotImplementedError(f"Only dihedral function 2 and 9 are supported in {section} of {top.source}: {raw_line}")

    return DihedralType(ai, aj, ak, al, funct, params)


def _parse_atom(parts: List[str], mol: MoleculeType, fn_source: str | Path) -> Atom:
    raw_line = " ".join(parts)
    section = "atoms"

    _require(parts, 7, section=section, source=fn_source, raw=raw_line)
    nr = int(parts[0])
    type = parts[1]
    resnr = int(parts[2])
    residue = parts[3]
    atom = parts[4]
    cgnr = int(parts[5])
    charge = float(parts[6])
    mass = float(parts[7]) if len(parts) >= 8 else None

    if any(a.nr == nr for a in mol.atoms):
        raise ValueError(f"Duplicate atom nr={nr} in moleculetype '{mol.name}' ({fn_source}): {raw_line}")

    return Atom(nr, type, resnr, residue, atom, cgnr, charge, mass)


def _parse_bond(parts: List[str], top: Topology, mol: MoleculeType) -> Bond:
    raw_line = " ".join(parts)
    section = "bonds"

    _require(parts, 3, section=section, source=top.source, raw=raw_line)
    ai = _get_atom_by_nr(mol, int(parts[0]))
    aj = _get_atom_by_nr(mol, int(parts[1]))
    funct = int(parts[2])

    if len(parts) > 3:
        _require(parts, 5, section="bond", source=top.source, raw=raw_line)
        params = {'b0': float(parts[3]), 'kb': float(parts[4])}
    
    elif len(parts) == 3:
        # Try to get parameters from bondtypes
        bt = _lookup_paramtype(top, ai, aj)
        if bt is None:
            raise ValueError(f"Bond parameters not found for bond between atom {ai} and atom {aj} in {section} of {top.source}: {raw_line}")
        if bt.func != funct:
            raise ValueError(f"Bond function mismatch for bond between atom {ai} and atom {aj} in {section} of {top.source}: {raw_line}")
        params = bt.params

    else:
        raise ValueError(f"Invalid bond line in {top.source}: {raw_line}")
    
    return Bond(ai, aj, funct, params)
    

def _parse_pair(parts: List[str], top: Topology, mol: MoleculeType) -> Pair:

    raw_line = " ".join(parts)
    section = "pairs"

    _require(parts, 3, section=section, source=top.source, raw=raw_line)
    ai = _get_atom_by_nr(mol, int(parts[0]))
    aj = _get_atom_by_nr(mol, int(parts[1]))
    funct = int(parts[2])

    if len(parts) > 3:
        if funct == 1:
            _require(parts, 5, section="pair", source=top.source, raw=raw_line)
            params = {'sigma': float(parts[3]), 'epsilon': float(parts[4])}
        elif funct == 2:
            _require(parts, 8, section="pair", source=top.source, raw=raw_line)
            params = {
                'fudgeQQ': float(parts[3]),
                'q1': float(parts[4]),
                'q2': float(parts[5]),
                'sigma': float(parts[6]),
                'epsilon': float(parts[7]),
            }
    
    elif len(parts) == 3:
        # Try to get parameters from pairtypes
        pt = _lookup_paramtype(top, ai, aj, pair=True)
        if pt is None:
            raise ValueError(f"Pair parameters not found for pair between atom {ai} and atom {aj} in {section} of {top.source}: {raw_line}")
        if pt.func != funct:
            raise ValueError(f"Pair function mismatch for pair between atom {ai} and atom {aj} in {section} of {top.source}: {raw_line}")

        params = pt.params
    else:
        raise ValueError(f"Invalid pair line in {top.source}: {raw_line}")
    
    return Pair(ai, aj, funct, params)
    

def _parse_angle(parts: List[str], top: Topology, mol: MoleculeType) -> Angle:

    raw_line = " ".join(parts)
    section = "angles"
    _require(parts, 4, section=section, source=top.source, raw=raw_line)
    ai = _get_atom_by_nr(mol, int(parts[0]))
    aj = _get_atom_by_nr(mol, int(parts[1]))
    ak = _get_atom_by_nr(mol, int(parts[2]))
    funct = int(parts[3])

    if len(parts) > 4:
        if funct == 1:
            _require(parts, 6, section="angle", source=top.source, raw=raw_line)
            params = {'th0': float(parts[4]), 'kth': float(parts[5])}
        elif funct == 5:
            _require(parts, 8, section="angle", source=top.source, raw=raw_line)
            params = {
                'th0': float(parts[4]),
                'kth': float(parts[5]),
                's0': float(parts[6]),
                'kub': float(parts[7]),
            }
    elif len(parts) == 4:
        # Try to get parameters from angletypes
        at = _lookup_paramtype(top, ai, aj, ak)
        if at is None:
            raise ValueError(f"Angle parameters not found for angle between atom {ai}, {aj}, and {ak} in {section} of {top.source}: {raw_line}")
        if at.func != funct:
            raise ValueError(f"Angle function mismatch for angle between atom {ai}, {aj}, and {ak} in {section} of {top.source}: {raw_line}")

        params = at.params

    else:
        raise ValueError(f"Invalid angle line in {top.source}: {raw_line}")
    
    return Angle(ai, aj, ak, funct, params)


def _parse_dihedral(parts: List[str], top: Topology, mol: MoleculeType) -> Dihedral:

    raw_line = " ".join(parts)
    section = "dihedrals"
    _require(parts, 5, section=section, source=top.source, raw=raw_line)
    ai = _get_atom_by_nr(mol, int(parts[0]))
    aj = _get_atom_by_nr(mol, int(parts[1]))
    ak = _get_atom_by_nr(mol, int(parts[2]))
    al = _get_atom_by_nr(mol, int(parts[3]))
    funct = int(parts[4])
    if len(parts) > 5:
        if funct == 2:
            _require(parts, 7, section="dihedral", source=top.source, raw=raw_line)
            params = {'xi': float(parts[5]), 'kxi': float(parts[6])}
        elif funct == 9:
            _require(parts, 8, section="dihedral", source=top.source, raw=raw_line)
            params = {'phi0': float(parts[5]), 'kphi': float(parts[6]), 'mult': int(parts[7])}
    elif len(parts) == 5:
        # Try to get parameters from dihedraltypes
        dt = _lookup_paramtype(top, ai, aj, ak, al)
        if dt is None:
            raise ValueError(f"Dihedral parameters not found for dihedral between atom {ai}, {aj}, {ak}, and {al} in {section} of {top.source}: {raw_line}")
        if dt.func != funct:
            raise ValueError(f"Dihedral function mismatch for dihedral between atom {ai}, {aj}, {ak}, and {al} in {section} of {top.source}: {raw_line}")

        params = dt.params

    else:
        raise ValueError(f"Invalid dihedral line in {top.source}: {raw_line}")
    
    return Dihedral(ai, aj, ak, al, funct, params)


def _parse_exclusion(parts: List[str], top: Topology, mol: MoleculeType) -> Exclusion:
    
    raw_line = " ".join(parts)
    _require(parts, 1, section="exclusions", source=top.source, raw=raw_line)
    excluded = tuple(int(x) for x in parts)
    return Exclusion(excluded)


def _parse_settle(parts: List[str], top: Topology, mol: MoleculeType) -> Settle:
    raw_line = " ".join(parts)
    section = "settles"

    _require(parts, 4, section=section, source=top.source, raw=raw_line)
    ai = int(parts[0])
    func = int(parts[1])
    doh = float(parts[2])
    dhh = float(parts[3])
    return Settle(ai, func, doh, dhh)


def _parse_vsite3(parts: List[str], top: Topology, mol: MoleculeType) -> VirtualSite3:

    raw_line = " ".join(parts)
    section = "virtual_sites3"
    _require(parts, 5, section=section, source=top.source, raw=raw_line)

    ai = _get_atom_by_nr(mol, int(parts[0]))
    aj = _get_atom_by_nr(mol, int(parts[1]))
    ak = _get_atom_by_nr(mol, int(parts[2]))
    al = _get_atom_by_nr(mol, int(parts[3]))
    funct = int(parts[3])

    if funct == 1:
        _require(parts, 7, section=section, source=top.source, raw=raw_line)
        params = {'a': float(parts[5]), 'b': float(parts[6])}
    elif funct == 2:
        _require(parts, 7, section=section, source=top.source, raw=raw_line)
        params = {'a': float(parts[5]), 'd': float(parts[6])}
    elif funct == 3:
        _require(parts, 7, section=section, source=top.source, raw=raw_line)
        params = {'th': float(parts[5]), 'd': float(parts[6])}
    elif funct == 4:
        _require(parts, 8, section=section, source=top.source, raw=raw_line)
        params = {'a': float(parts[5]), 'b': float(parts[6]), 'c': float(parts[7])}
    else:
        raise NotImplementedError(f"Only virtual_sites3 function 1-4 are supported in {section} of {top.source}: {raw_line}")
    
    return VirtualSite3(ai, aj, ak, al, funct, params)


# -----------------------------
# Public API
# -----------------------------

def read_topology(fn: Union[str, Path]) -> Topology:
    fn = Path(fn).resolve()
    top = Topology(source=fn)
    _read_file(fn, top)
    return top


# -----------------------------
# Reader core
# -----------------------------

def _read_file(fn: Path, top: Topology) -> None:
    current_section: Optional[str] = None
    active_mol: Optional[MoleculeType] = None

    with fn.open() as f:
        for raw_line in f:
            raw_line = raw_line.rstrip("\n")
            line = _strip_comment(raw_line)

            if not line:
                continue

            # Preprocessor directives
            if line.startswith("#"):
                if line.startswith("#include"):
                    inc = line.split()[1].strip('"')
                    inc_path = (fn.parent / inc).resolve()
                    _read_file(inc_path, top)
                continue

            # Section header
            if line.startswith("["):
                current_section = line.strip("[]").strip()
                continue

            if not current_section:
                continue

            parts = line.split()

            # ---- global sections ----
            if current_section == "defaults":
                _require(parts, 5, section="defaults", source=fn, raw=raw_line)
                top.defaults = Defaults(
                    nbfunc=int(parts[0]),
                    comb_rule=int(parts[1]),
                    gen_pairs=parts[2],
                    fudgeLJ=float(parts[3]),
                    fudgeQQ=float(parts[4]),
                )
                continue

            if current_section == "atomtypes":
                atomtype = _parse_atomtype(parts, top)
                top.atomtypes.append(atomtype)
                continue

            if current_section == "bondtypes":
                bondtype = _parse_bondtype(parts, top)
                top.bondtypes.append(bondtype)
                continue

            if current_section == "pairtypes":
                pairtype = _parse_pairtype(parts, top)
                top.pairtypes.append(pairtype)
                continue

            if current_section == "angletypes":
                angletype = _parse_angletype(parts, top)
                top.angletypes.append(angletype)
                continue

            if current_section == "dihedraltypes":
                dihedraltype = _parse_dihedraltype(parts, top)
                top.dihedraltypes.append(dihedraltype)
                continue

            if current_section == "system":
                # often one-line free text
                top.system = raw_line.strip()
                continue

            if current_section == "molecules":
                _require(parts, 2, section="molecules", source=fn, raw=raw_line)
                molname = parts[0]
                mol = top.moleculetypes.get(molname)
                count = int(parts[1])
                top.molecules[molname] = (mol, count)
                continue

            # ---- molecule definition ----
            if current_section == "moleculetype":
                _require(parts, 2, section="moleculetype", source=fn, raw=raw_line)
                name, nrexcl = parts[0], int(parts[1])
                if name in top.moleculetypes:
                    raise ValueError(f"Duplicate moleculetype '{name}' in {fn}: {raw_line}")
                active_mol = MoleculeType(name=name, nrexcl=nrexcl)
                top.moleculetypes[name] = active_mol
                continue

            # From here on, we need an active molecule
            # if active_mol is None:
            #     # Ignore stray molecule-scoped content if topology is malformed,
            #     # but usually this should be an error.
            #     raise ValueError(f"Section [ {current_section} ] appears before any [ moleculetype ] in {fn}: {raw_line}")

            if current_section == "atoms":
                atom = _parse_atom(parts, active_mol, fn)
                active_mol.atoms.append(atom)
                continue

            if current_section == "bonds":
                bond = _parse_bond(parts, top, active_mol)
                active_mol.bonds.append(bond)
                continue

            if current_section == "pairs":
                pair = _parse_pair(parts, top, active_mol)
                active_mol.pairs.append(pair)
                continue

            if current_section == "angles":
                angle = _parse_angle(parts, top, active_mol)
                active_mol.angles.append(angle)
                continue

            if current_section == "dihedrals":
                dihedral = _parse_dihedral(parts, top, active_mol)
                active_mol.dihedrals.append(dihedral)
                continue

            if current_section == "exclusions":
                exclusion = _parse_exclusion(parts, top, active_mol)
                active_mol.exclusions.append(exclusion)
                continue

            if current_section == "settles":
                settle = _parse_settle(parts, top, active_mol)
                active_mol.settles.append(settle)
                continue

            if current_section in ["virtual_sites3", "dummies3"]:
                vsite = _parse_vsite3(parts, top, active_mol)
                active_mol.virtual_sites3.append(vsite)
                continue

            # Unsupported section in this parser: ignore.
            continue
```
